```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
features=['Intelligence', 'Strength', 'Speed', 'Durability', 'Power', 'Combat']
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(14, 8), gridspec_kw={'hspace': 0.5, 'wspace': 0.3})

for i, ax in enumerate(axes.flatten()):
    sns.kdeplot(x=df[df.Alignment=='good'][features[i]], ax=ax)
    sns.kdeplot(x=df[df.Alignment=='bad'][features[i]], ax=ax)
    ax.set_title(features[i])
    ax.legend(['good', 'bad'])
    
# Add a title for the whole figure
fig.suptitle('Density plots of six features')

# Display the plot
plt.show()
```
